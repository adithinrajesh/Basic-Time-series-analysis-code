Experiment 5 - To check if a given time series is seasonal or has stationarity.
1 - Seasonality
    a - Plot acf with and without lag(3)
    b - p value test
2 - Stationarity
    a - Plot histogram
    b - Compute mean and variance
    c - ADF test

Seasonality

```python
#Importing libraries
import pandas as pd
from matplotlib import pyplot
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import acf
```

```python
#Plotting
series = pd.read_csv("finaldata.csv")
plot_acf(series)
pyplot.show()
print()
plot_acf(series, lags = 3)
pyplot.show()
```

```python
#Printing pvalue
func, confidence_interval, qstat, pvalue = acf(series, nlags = 3)
print(pvalue)
#p value less than 0.05 is seasonal
```

Stationarity

```python
#histogram
import pandas as pd
import matplotlib.pyplot as plt
series1 = pd.read_csv("daily-total-female-births.csv", header=0, index_col=0)
series1.hist()
plt.show()
```

```python
#computing mean and variance
val = series1.values
split = round(len(val)/2)
split_1, split_2 = val[0:split], val[split:]
split_1_var, split_2_var = split_1.var(), split_2.var()
split_1_mean, split_2_mean = split_1.mean(), split_2.mean()
print("Mean 1 = ", split_1_mean, " and Mean 2 = ", split_2_mean)
print("Variance 1 = ", split_2_var, " and Variance 2 = ", split_2_var)
#since the values mean and var are close to each other series is stationary
```

```python
#Dicky fuller test for checking stationarity
import pandas as pd
from statsmodels.tsa.stattools import adfuller
#series1 is used here
val = series1.values
result = adfuller(val)
print("ADF statistic = ", result[0])
print("p-value = ", result[1])
#if p value is less than or equal to 0.05 reject null hypo, series is stationary.
```
